```python
import pandas as pd
df = pd.read_csv('/content/netflix_titles.csv')
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df['director'].fillna('Unknown', inplace=True)
```

```python
df['cast'].fillna('Not_Listed', inplace=True)
```

```python
df['country'].fillna(df['country'].mode()[0], inplace=True)
```

```python
df
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], format='mixed')
df
```

```python
df['add_year'] = df['date_added'].dt.year
df['add_month'] = df['date_added'].dt.month
```

```python
df
```

```python
# how many movies and tv shows
df['type'].value_counts()
```

```python
# top 10 counttries with most content
df['country'].value_counts().head(10)
```

```python
# most common genre listed in
df['listed_in'].value_counts().head(3)
```

```python
# most frequesnt directots
df['director'].value_counts().head()
```

```python
# content released each year
df['release_year'].value_counts().sort_index()
```

```python
# shows added in each month
df.groupby(['add_year', 'add_month'])['show_id'].size()
```

```python
# How long is the longest movie (duration)?
filt_movies = df['type'] == 'Movie'
movies = df.loc[filt_movies, ['type', 'duration']]
df['duration_int'] = movies['duration'].str.replace('min', ' ').astype(float)
max_duration = df['duration_int'].max()
filt_max_duration = df['duration_int'] == max_duration
df.loc[filt_max_duration, ['type', 'duration']]
```

```python
# shows were added per year.
df.groupby('add_year')['show_id'].size()
```

```python
# director has made the most TV Shows.
TV_SHOWS = df[df['type'] == 'TV Show']
TV_SHOWS = TV_SHOWS[TV_SHOWS['director']!='Unknown']
TV_SHOWS['director'].value_counts().head(1)
```

```python
# movies are longer than 2 hours (if duration is in minutes).
movies = df[df['type'] == 'Movie']
more_2hrs = movies[movies['duration_int'] > 120]
more_2hrs[['type', 'duration']]
```

```python
# age-based rating categories from the rating column.
def age_grp(x):
  if x in ['TV-Y', 'TV-G']:
      return 'Kids'
  elif x in ['PG', 'TV-PG']:
      return 'Older Kids'
  elif x in ['PG-13', 'TV-14']:
      return 'Teens'
  elif x in ['R', 'TV-MA', 'NC-17']:
      return 'Adults'
  else:
      return 'Unrated'
df['Age_group'] = df['rating'].apply(age_grp)
df[['rating', 'Age_group']]
```
